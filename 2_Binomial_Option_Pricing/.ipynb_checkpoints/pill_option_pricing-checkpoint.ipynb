### Monte Carlo Option Pricing & Black-Scholes-Merton

```python
# Import of libraries
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

Definition of functions - namely the Black-Scholes-Merton pricing of European call and put options and their payoffs.

```python
def BSM_European_Call(S_0, K, r, sigma, T):
    d1 =  (1/(sigma * np.sqrt(T))) * (np.log(S_0/K) + (r + 0.5 * sigma ** 2) * T)
    d2 = d1 - sigma * np.sqrt(T) 
    return S_0 * norm.cdf(d1) - K * np.exp(-r*T) * norm.cdf(d2)

def BSM_European_Put(S_0, K, r, sigma, T): # This uses the Put-Call parity
    d1 =  (1/(sigma * np.sqrt(T))) * (np.log(S_0/K) + (r + 0.5 * sigma ** 2) * T)
    d2 = d1 - sigma * np.sqrt(T) 
    return - S_0 *(1 - norm.cdf(d1)) + K * np.exp(-r*T) *(1- norm.cdf(d2))

def Payoff_Call(S_T, K):
    return np.maximum(S_T - K, 0)

def Payoff_Put(S_T, K):
    return np.maximum(K - S_T, 0)
```

Function that runs Monte Carlo simulations - It runs $R$ independent realizations of a Binomial tree with risk free probability measure (Binomial) $q = (e^{r\Delta t} - d)/(u-d)$. It returns prices and standard deviations.

```python
def Call_Put_MonteCarlo(R, S_0, K, r, sigma, T, N):
    # Cox - Ross- Rubenstein parameters for the Binomial tree
    dt = T/N
    u = np.exp(sigma*np.sqrt(dt))
    d = 1/u
    p = (np.exp(r * dt) - d)/(u - d)
    
    # Preparation of the samples ...
    x = np.zeros((R, N)) # Creating price evolution array
    x[:,0] = S_0 # initial price of the asset 

    # This runs the Binomial tree (MC, prices)
    for t in range(1,N):
        trial = np.random.uniform(size=R)
        x[trial<=p,t] = u * x[trial<=p,t-1]   # If trial <= p then asset price goes up
        x[trial>p ,t] = d * x[trial>p ,t-1]   # If trial  > p then asset price goes down 
    
    # Simulation done
    # Time to compute average pay off and discounting it
    call_payoff = np.maximum(x[:,-1]-K,0) # Computes array of pay offs (Call)
    put_payoff = np.maximum(K-x[:,-1],0)  # Computes array of pay offs (Put)
    disc_call_payoff = np.exp(- r * N * dt) * call_payoff
    disc_put_payoff = np.exp(- r * N * dt) * put_payoff

    # The average if the actual price.
    c = np.mean(disc_call_payoff, axis = 0)
    p = np.mean(disc_put_payoff, axis = 0)

    std_c = np.std(disc_call_payoff, axis = 0)
    std_p = np.std(disc_put_payoff, axis = 0)

    return c, p, std_c, std_p
```

Example of use

```python
c_mc,p_mc, _, _ = Call_Put_MonteCarlo(10**4, 1, 1.2, 0.1, 0.2, 1, 10**3)
c = BSM_European_Call(1, 1.2, 0.1, 0.2, 1)
p = BSM_European_Put(1, 1.2, 0.1, 0.2, 1)
```

```python
print(f"Monte Carlo simulation: Call {c_mc} Put {p_mc}")
print(f"Actual values BSM: Call {c} Put {p}")
```

#### Convergence towards the BSM pricing.

We study here the convergence towards the Black-Scholes-Merton (BSM) pricing of European options as the size of the tree $n$ increases - remember, how the Cox-Ross-Rubenstein trees converge towards geometric Brownian motion (GBM) as $\Delta t$ becomes small.

```python
# Paramter values
S_0 = 1.0
K = 1.2
r = 0.05
sigma = 0.1
T = 1.0
R = 10**4

N_list = [10, 50, 100, 500, 1000] # Fixed T different N means different Dt
call_reference = BSM_European_Call(S_0, K, r, sigma, T)
c_N_values_results = []
for N in N_list:
    c, _, _, _ = Call_Put_MonteCarlo(R, S_0, K, r, sigma, T, N)
    c_N_values_results.append((N, c, c - call_reference))
```

```python
print(f"BSM price (normalized): {call_reference:.10f}\n")
print(" n       c(MC)       c(MC)-c(BSM)")
for n,c,diff in c_N_values_results:
    print(f"{n:4d}  {c:12.10f}  {diff: .10f}")
```

We plot now the convergence to make it look clearer, as the size $n$ increases.

```python
# Paramter values
S_0 = 1.0
K = 1.2
r = 0.05
sigma = 0.2
T = 1.0
R = 10**4

call_reference = BSM_European_Call(S_0, K, r, sigma, T)
N_array = np.arange(1, 100, 2)

c_mc_vals = []
std_c_mc_vals = []
for N in N_array:
    c, _, std_c, _ = Call_Put_MonteCarlo(R, S_0, K, r, sigma, T, N)
    c_mc_vals.append(c)
    std_c_mc_vals.append(std_c)
```

```python
c_mc_vals_array = np.array(c_mc_vals)
std_c_mc_vals_array = np.array(std_c_mc_vals)

plt.figure(figsize=(4,3))
plt.plot(N_array, c_mc_vals, marker = 'o', label = 'CRR MC ($N$)', alpha = 0.5, color = 'blue')
plt.fill_between(N_array, c_mc_vals_array - 0.1*std_c_mc_vals_array, c_mc_vals_array + 0.1*std_c_mc_vals_array, color="blue", alpha=0.3)#, label="±1σ")

plt.hlines(y=call_reference, xmin=-1, xmax=100, linestyle = '--', color = 'grey', label = 'BSM')

plt.xlim([-1,100])
plt.xlabel(r'$N$', fontsize=12)
plt.ylabel(r'$c$', fontsize=12)
plt.legend(loc = 'lower right', frameon = False)
plt.savefig('N_convergence.pdf', bbox_inches='tight')
plt.show()
```

#### Predictive capacity for large enough trees ($N \rightarrow \infty$)

We take now $N=10^3$ and compute for different values of K the value of a European Call and Put option, comparing with the BSM pricing.

```python
# Paramter values
S_0 = 1.0
r = 0.05
sigma = 0.1
T = 1.0
R = 10**4
N = 10**3

# List of values of the strike price K
K_list = np.linspace(0.6, 1.4, 20)
```

```python
c_vals = []
p_vals = []

for K in K_list:
    c, p, std_c, std_p = Call_Put_MonteCarlo(R, S_0, K, r, sigma, T, N)
    c_vals.append(c)
    p_vals.append(p)
```

```python
plt.figure(figsize=(4,3))

plt.plot(K_list, c_vals, marker = 'o', linestyle = 'none', alpha = 0.5, label = 'CRR MC (Call)')
plt.plot(K_list, p_vals, marker = 'o', linestyle = 'none', alpha = 0.5, label = 'CRR MC (Put)')

# Plot BSM values 
K_range = np.linspace(0.6, 1.4, 100)
c_bsm = BSM_European_Call(S_0, K_range, r, sigma, T)
p_bsm = BSM_European_Put(S_0, K_range, r, sigma, T)
plt.plot(K_range, c_bsm, marker = 'none', linestyle = '--', color = 'black', label = 'BSM Eur. Call')
plt.plot(K_range, p_bsm, marker = 'none', linestyle = ':', color = 'black', label = 'BSM Eur. Put')


plt.legend(loc = 'upper center', frameon=True)#, bbox_to_anchor=(1.525,1.025))
plt.xlabel(r'$K$', fontsize = 12)
plt.ylabel(r'$c, p$', fontsize = 12)
plt.savefig('simulations_and_prediction.pdf', bbox_inches='tight')
plt.show()
```

We now perform simulations for a range of paramters $(K,r)$ and draw the prediction using BSM alongside.

```python
# Paramter values
S_0 = 1.0
#r = 0.05
sigma = 0.1
T = 1.0
R = 10**4
N = 10**3

# List of values of the strike price K
K_list = np.linspace(0.6, 1.4, 20)
r_list = np.linspace(0,0.1, 20)
```

```python
c_vals_mat = []
p_vals_mat = []

for r in r_list:
    c_vals = []
    p_vals = []
    r_index = np.where(r_list == r)[0]
    
    for K in K_list:
        c, p, std_c, std_p = Call_Put_MonteCarlo(R, S_0, K, r, sigma, T, N)
        c_vals.append(c)
        p_vals.append(p)

    # Append now vector to matrix
    c_vals_mat.append(c_vals)
    p_vals_mat.append(p_vals)

    print(f" Progress : {r_index+1}/{len(r_list)}")

print("Done")
```

Save the generated data.

```python
# Guardar
np.savez("montecarlo_results.npz",
         c_vals_mat=np.array(c_vals_mat),
         p_vals_mat=np.array(p_vals_mat),
         r_list=np.array(r_list),
         K_list=np.array(K_list))
```

Read the data and plot it.

```python
# Cargar
data = np.load("montecarlo_results.npz")
c_vals_mat = data["c_vals_mat"]
p_vals_mat = data["p_vals_mat"]
r_list = data["r_list"]
K_list = data["K_list"]
```
