# ðŸ“˜ Introduction

Geometric Brownian Motion (GBM) is one of the most fundamental stochastic processes in quantitative finance. It models the evolution of asset prices by incorporating both deterministic trends and random fluctuations, making it a natural candidate for describing stock price dynamics under uncertainty.

Mathematically, GBM is governed by the stochastic differential equation (SDE)

\begin{equation}
    dS_t = \mu S_t \; dt + \sigma S_t \; dW_t
\end{equation}

where $S_t$ is the asset price at time $t$, $\mu $ is the drift (average return), $\sigma$ is the volatility (standard deviation of returns) and $W_t$ is a standard Wiener process (Brownian motion).

This notebook explores GBM from three complementary angles:

- [1. Numerical Simulation via SDE Integration](#1-numerical-simulation-via-sde-integration) We simulate GBM paths using the Eulerâ€“Maruyama method and analyze the statistical properties of the resulting log-returns. We show how, after normalization, they collapse into a standard normal distribution as theory predicts.
- [2. Binomial and Monte Carlo Simulation](#2-binomial-and-monte-carlo-simulation) We build a discrete-time binomial approximation of GBM and simulate multiple price trajectories using Monte Carlo methods. This section focuses on statistical convergence and the central limit behavior of log-returns.
- [3. Real Market Data Analysis](#3-real-market-data-analysis) We retrieve historical price data from financial markets and examine to what extent real asset returns align with the assumptions and predictions of GBM. This empirical perspective helps highlight both the strengths and limitations of the model.

## 1. Numerical Simulation via SDE Integration


Geometric Brownian Motion (GBM) is defined by the stochastic differential equation:

\begin{equation}
dS_t = \mu S_t \, dt + \sigma S_t \, dW_t
\end{equation}

As shown in the notes, the logarithm of prices follows a normal distribution:

\begin{equation}
\log\left(\frac{S_t}{S_0}\right) \sim \mathcal{N}\left( \left(\mu - \frac{1}{2} \sigma^2\right)t, \, \sigma^2 t \right)
\end{equation}

In this section, we simulate GBM numerically using the Eulerâ€“Maruyama method, a first-order discretization scheme for SDEs. We generate sample paths and study the distribution of log-returns:

\begin{equation}
r_t = \log\left(\frac{S_t}{S_0}\right)
\end{equation}

To verify the theoretical distribution, we normalize the log-returns to construct:

\begin{equation}
Z = \frac{r_t - \left(\mu - \frac{1}{2} \sigma^2\right)t}{\sigma \sqrt{t}} \sim \mathcal{N}(0, 1)
\end{equation}

By plotting histograms and comparing with the standard normal density, we test the consistency between our numerical simulations and the analytic solution.

```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib.gridspec import GridSpec
```

```python
# Parameters 
mu = 0.2
sigma = 0.1

N = 10**5  # Number of trajectories
T = 1000   # Number of timesteps
dt = 0.01 # dt
time_vec = np.linspace(0,T*dt,1000)
```

```python
def log_normal(x,mu,sigma,t): # not the original, but the one for the GBM, provide mu, sigma and t
    prefactor = 1/(x * sigma * np.sqrt(2*np.pi*t) )
    num = (np.log(x) - (mu-0.5*sigma**2)*t)**2
    denom = 2 * sigma**2 * t
    f = prefactor * np. exp(-num/denom)
    return f
```

```python
# Generation of trajectories
x = np.zeros((N, T)) # Generating the arrows
x[:,0] = 1 # initial condition
time = np.arange(0,T*dt,dt)

for t in range(1,T):
    x[:,t] = x[:,t-1] * (1 + dt * mu + sigma * np.random.normal(0, np.sqrt(dt), size=N))
```

```python
# Two column figure: trajectories + horizontal histogram
fig = plt.figure(figsize=(7, 3))
gs = GridSpec(1, 2, width_ratios=[2, 1], wspace=0.2)

# Left pannel : trajectories
ax0 = fig.add_subplot(gs[0])
for i in range(250): # Take only 250
    ax0.plot(time, x[i, :], marker='none', alpha=0.05)
    
st_average = np.exp(mu*time_vec)
ax0.plot(time_vec, st_average, linestyle='--', color='black', label=r'$e^{\mu t}$')
ax0.axvline(x=10, ymin=0, ymax= 20, linestyle = '--', color = 'grey')
ax0.set_xlabel(r'$t$')
ax0.set_ylabel(r'$S_t$')
ax0.set_ylim([0,20])
ax0.text(9.5, 15, r'$t = T$', rotation='vertical')
ax0.set_yticks([0, 5, 10, 15, 20])
ax0.legend(loc='best', frameon=False)

# Right pannel: histogram
ax1 = fig.add_subplot(gs[1], sharey=ax0)
ax1.hist(x[:, -1], bins=50, density=True, orientation='horizontal', color='steelblue', edgecolor = 'white', label = r'Hist $(S_t)$')
t_teo = np.linspace(0.1,20,1000)
pdf_t = log_normal(t_teo,mu,sigma,10)
ax1.plot(pdf_t, t_teo, color = 'black', label = r'$\ln S_T \sim N(\mu_{T}, \sigma_{T}^2)$')
ax1.set_xlabel(r'$pdf(S_T)$')
ax1.set_ylim([0,20])
ax1.legend(loc = 'best', frameon=False)

#plt.tight_layout()
plt.show()
```

So far, we have seen that the log-normal model (using $\mu$ and $\sigma$, rather than $\mu_t$ and $\sigma_t$) provides a good fit to the histograms obtained from the numerical integration of the SDE. We will now use our theoretical insights to collapse the data onto a single curve.

```python
# Re-definition of parameters
mu_list = [0.1, 0.2, 0.3] # list of values for mu
sigma = 0.1

N = 10**5  # Number of trajectories
T = 1000   # Number of timesteps
dt = 0.01 # dt
```

```python
# Generation of data.
x_T_matrix = [] # Keep data for each mu here (last step, t = T*dt)

for mu_i  in mu_list:
    # Generation of trajectories
    x = np.zeros((N, T)) # Generating the arrows
    x[:,0] = 1 # initial condition
    time = np.arange(0,T*dt,dt)

    for t in range(1,T):
        x[:,t] = x[:,t-1] * (1 + dt * mu_i + sigma * np.random.normal(0, np.sqrt(dt), size=N))

    # Append data of the last time step, at time t = T*dt
    x_T_matrix.append(x[:,-1])

print("Done")
```

```python
print(x_T_matrix[1])
```

```python
plt.figure(figsize=(4, 3))

for i in range(len(x_T_matrix)):
    # Plotting histogram(s)
    x_T = x_T_matrix[i]
    plt.hist(x_T, bins = 50, density=True, label = r'$\mu= $ '+str(mu_list[i]), edgecolor = 'white')

plt.xlabel(r'$S_T$')
plt.ylabel(r'pdf$(S_T)$')
# plt.yscale('log')
plt.xscale('log')
plt.legend(loc='best', frameon = False)
plt.show()
```

```python
plt.figure(figsize=(4, 3))

for i in range(len(x_T_matrix)):
    # Plotting histogram
    x_T = x_T_matrix[i]
    z_T = (np.log(x_T) - (mu_list[i] -0.5*sigma**2)*T*dt )/(sigma*np.sqrt(T*dt)) 

    # Counting frequencies.
    counts, bin_edges = np.histogram(z_T, bins=50, density=True)
    # Compute center of bins
    bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
    plt.plot(bin_centers, counts, 'o', label=r'$\mu=$ '+str(mu_list[i]), linestyle = 'none', alpha = 0.25)

# Plotting N(0,1)
z_gauss = np.linspace(-4,4, 500)
normal_gauss = (1/(np.sqrt(2*np.pi)))*np.exp(-0.5*(z_gauss**2))
plt.plot(z_gauss, normal_gauss, color = 'black', linestyle = '--', label = r'$N(0,1)$')

plt.xlabel(r'$X_T = \frac{\ln S_T - (\mu - \sigma^2/2)T}{\sigma \sqrt{T}}$')
plt.ylabel(r'pdf$(Z_T)$')
plt.yscale('log')
plt.legend(loc='best', frameon = False)
plt.show()
```

## 2. Binomial and Monte Carlo Simulation


An alternative to solving the SDE directly is to approximate GBM via a binomial model. In this discrete-time approach, the asset price evolves according to:

\begin{equation}
S_{n+1} = 
\begin{cases}
 uS_n & \text{with probability } p \\
 dS_n & \text{with probability } 1 - p
\end{cases}
\end{equation}

with $u = e^{\sigma \sqrt{\Delta t}}$, $d = e^{-\sigma \sqrt{\Delta t}}$, and $\Delta t$ being the time increment. The binomial model converges to GBM as $\Delta t \to 0$, in accordance with the central limit theorem.

In this section, we implement Monte Carlo simulations of this binomial approximation. By generating a large number of independent paths, we analyze:

- The convergence of the empirical distribution of log-returns
- The behavior of the mean and variance as a function of time
- The agreement with the theoretical normal distribution of $\log(S_t / S_0)$

This method highlights the probabilistic foundations of GBM and provides a natural bridge between discrete-time and continuous-time modeling in finance.


```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib.gridspec import GridSpec
```

```python
# Properties of the change regarding stochasticity of price of asset
mu = 0.1
sigma = 0.1
dt = 0.001

# Properties of the dynamics and statistics
N = 10000 # Number of stochastic steps
R = 10000  # Numbre of repetitions
time = np.linspace(0,N*dt,N) # time vector

u = np.exp(sigma*np.sqrt(dt)) # Paramters of the tree (Cox-Ross-Rubinstein)
d = 1/u
p = (np.exp(mu * dt) - d)/(u-d)
```

```python
# Preparation of the samples ...
x = np.zeros((R, N)) # Generating the arrows
x[:,0] = 1 # initial price of the asset S_0 = 1

for t in range(1,N):
    trial = np.random.uniform(size=R)
    x[trial<=p,t] = u * x[trial<=p,t-1]   # If trial <= p then asset price goes up
    x[trial>p ,t] = d * x[trial>p ,t-1]   # If trial  > p then asset price goes down 

print("Done")
```

```python
# Three column figure: trajectories + horizontal histogram
fig = plt.figure(figsize=(10, 3))
gs = GridSpec(1, 3, width_ratios=[1, 1, 1], wspace=0.2)

# Left pannel: trajectories
ax0 = fig.add_subplot(gs[0])
for i in range(0,100):
    ax0.plot(x[i,:], alpha = 0.05, marker = 'o')

ax0.set_ylim([d**10,u**10])
ax0.set_xlim([0,10])
ax0.set_xlabel(r'$t/dt$ (it.)')
ax0.set_ylabel(r'$S_t$')

# Central pannel: histogram
ax1 = fig.add_subplot(gs[1])
for i in range(0,100):
    ax1.plot(time, x[i,:], marker = 'none', alpha = 0.1)

# Average trajectory
st_avg = np.mean(x, axis=0)
ax1.plot(time, st_avg, linestyle = '--', color = 'black', label = r'$\langle S_t \rangle$')

ax1.set_xlabel(r'$t$')
ax1.legend(loc = 'best', frameon=False)
ax1.set_ylim([0,8])

# Right pannel: histogram
ax2 = fig.add_subplot(gs[2])
ax2.hist(x[:, -1], bins=25, density=True, orientation='horizontal', color='steelblue', edgecolor = 'white', label = r'Hist $(S_T)$')
ax2.set_xlabel(r'pdf$(S_T)$')
ax2.legend(loc='best', frameon=False)
ax2.set_ylim([0,8])

#plt.tight_layout()
plt.show()
```

So far, we have seen how the discrete-time binomial approximation qualitatively captures the behavior of GBM.
```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib.gridspec import GridSpec
```

```python
# Properties of the change regarding stochasticity of price of asset
sigma = 0.1
dt = 0.001
mu_list = [0.1, 0.2, 0.3]

# Properties of the dynamics and statistics
N = 10000 # Number of stochastic steps
R = 10000  # Numbre of repetitions
time = np.linspace(0,N*dt,N) # time vector

u = np.exp(sigma*np.sqrt(dt))
d = 1/u
```

```python
x_T_list = [] # Keep for every mu the last value of each trajectory
averaged_trajectory = [] # Keep the averaged trajectory \sim e^{\mu t}?

# Generating the data
for mu in mu_list:
    p = (np.exp(mu*dt) - d)/(u-d) # Computing p using drift value
    # Preparation of the samples ...
    x = np.zeros((R, N)) # Generating the arrows
    x[:,0] = 1 # initial price of the asset S_0 = 1
    
    for t in range(1,N):
        trial = np.random.uniform(size=R)
        x[trial<=p,t] = u * x[trial<=p,t-1]   # If trial <= p then asset price goes up
        x[trial>p ,t] = d * x[trial>p ,t-1]   # If trial  > p then asset price goes down 

    # Append the last value of each trajectory to x_T_list
    x_T_list.append(x[:,-1])

    # Append averarged trajectory
    averaged_trajectory.append(np.mean(x, axis = 0))

print("Done")
```

```python
def log_normal(x,mu,sigma,t): # Again, not the original log-normal, adapted for GBM.
    prefactor = 1/(x * sigma * np.sqrt(2*np.pi*t) )
    num = (np.log(x) - (mu-0.5*sigma**2)*t)**2
    denom = 2 * sigma**2 * t
    f = prefactor * np. exp(-num/denom)
    return f
```

```python
plt.figure(figsize=(4, 3))

for i in range(len(mu_list)):
    plt.hist(x_T_list[i], bins=50, density=True, edgecolor = 'white', label = r'$\mu=$ '+str(mu_list[i]), alpha = 0.6)

# Now plotting the log normal distribution of the crap ...
st_range = np.linspace(0.1,40, 10000)
for i in range(len(mu_list)):
    mu = mu_list[i] # Take current value of drift mu
    pdf = log_normal(st_range, mu, sigma, N*dt)
    plt.plot(st_range, pdf, color = 'black')

plt.xlim([0,40])
plt.xlabel(r'$S_T$')
plt.ylabel(r'pdf$(S_t)$')
plt.legend(loc = 'best', frameon=False)
plt.show()
```

```python
plt.figure(figsize=(4, 3))

for i in range(len(mu_list)):
    # Plotting histogram
    x_T = x_T_list[i]
    z_T = (np.log(x_T) - (mu_list[i] -0.5*sigma**2)*N*dt )/(sigma*np.sqrt(N*dt)) 

    # Counting frequencies.
    counts, bin_edges = np.histogram(z_T, bins=50, density=True)
    # Calcular centros de los bins
    bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
    plt.plot(bin_centers, counts, 'o', label=r'$\mu=$ '+str(mu_list[i]), linestyle = 'none', alpha = 0.25)

# Plotting N(0,1)
z_gauss = np.linspace(-4,4, 500)
normal_gauss = (1/(np.sqrt(2*np.pi)))*np.exp(-0.5*(z_gauss**2))
plt.plot(z_gauss, normal_gauss, color = 'black', linestyle = '--', label = r'$N(0,1)$')

plt.xlabel(r'$X_T = \frac{\ln S_T - (\mu - \sigma^2/2)T}{\sigma \sqrt{T}}$')
plt.ylabel(r'pdf$(X_T)$')
plt.yscale('log')
plt.legend(loc='best', frameon = False)
plt.show()
```

## 3. Real Market Data Analysis


While GBM provides a mathematically elegant and tractable model for asset prices, it is ultimately an idealization. In this section, we confront the model with real financial data.

We use historical price data (e.g., from equities or indices) and compute the log-returns:

\begin{equation}
    r_t = \log \left( \frac{S_t}{S_{t-\Delta t}} \right)
\end{equation}

The core assumptions of GBM that we aim to test are:

- Log-returns are normally distributed
- The mean and variance are constant (i.e., stationary increments)
- Returns are independent over time (no autocorrelation)

We visualize the data using histograms, QQ plots, and autocorrelation functions. We also estimate the drift and volatility empirically:

\begin{equation}
\bar{r_t} = \frac{1}{T} \sum_{t} r_t, \quad \sigma_{r_t}^2 = \frac{1}{T-1} \sum_t (r_t - \bar{r_t})^2
\end{equation}

This section helps to understand in which regimes GBM is a good approximation â€” and where it falls short, for instance due to fat tails, volatility clustering, or leverage effects. These observations motivate the need for more sophisticated models in real financial modeling.


```python
# Uncom. to install Yahoo Finance
# !pip install yfinance
```

```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
import scipy.stats as stats
```

```python
# Downloading directly using YahooFinance (yfinance)
data = yf.download("GOOG", start="2020-01-01", end="2025-01-01")

# Showing the first rows
print(data.head())
```

We will only keep track of the "Closing" price of an asset (last value recorded on a given day).

```python
data_v = data["Close"]
print(data_v)
```

We save in into a CSV file to easily acces in the absence of internet connection.

```python
# Save the downloaded data to a CSV file
data_v.to_csv("GOOG_20-24_Close.csv")
print("Data saved to GOOG_20-24_Close.csv")
```

The downloaded data can be now read using pandas.

```python
# Reading the saved CSV file
data_df = pd.read_csv("GOOG_20-24_Close.csv", index_col=0, parse_dates=True)
data_df # Note the columns now are "Date" and "GOOG"
```

Using directly the data frame we can plot the time series of the evolution of the price of an asset.

```python
plt.figure(figsize=(4,3))

plt.plot(data_df, marker='none', label = 'GOOG')

plt.xlabel("Date")
plt.ylabel("Closing ($)")
plt.xticks(rotation=45)
plt.legend(loc = 'best', frameon=False)
plt.show()
```

We now compute the log returns with a time gap $\Delta t$ of a single day. This basically requires computing $r_t = \log (S_{t}/S_{t-1})$, where now $t$ is the date, and so $S_t$ the closing price of the asset on a given day $t$. These can be refered to ass "daily log return".

```python
# Computing the returns
log_returns_daily = np.log(data_df["GOOG"] / data_df["GOOG"].shift(1)).dropna() # Shift by one S_t -> S_{t-1}
```

```python
# Plotting the daily
plt.figure(figsize=(4,3))

plt.plot(log_returns_daily, marker='none', label = 'GOOG')

plt.xlabel("Date")
plt.ylabel(r"Daily $r_t$")
plt.xticks(rotation=45)
plt.legend(loc = 'best', frameon=True)
plt.show()
```

We can now draw a histogram to check how the daily log returns are distributed. The GBM predict that log returns for small $\Delta t$ are normally distributed as explained in the notes. We henece here compare the distribution of $r_t$ to a normal distribution with mean $\bar{r_t}$ and variane $\sigma_{r_{t}} = \sqrt{\text{Var}(r_t)}$.

```python
plt.figure(figsize=(4,3))

plt.hist(log_returns_daily, bins=50, density = True, label = 'GOOG', edgecolor = 'white')
x = np.linspace(log_returns_daily.min(), log_returns_daily.max(), 500)
plt.plot(x, stats.norm.pdf(x, log_returns_daily.mean(), log_returns_daily.std()), label="N(Î¼_t, Ïƒ_t)", color="black")
plt.legend(loc = 'best', fontsize = 10)
plt.xlabel(r'$r_t = \ln \left(\frac{S_{t}}{S_{t-1}}\right)$')
plt.ylabel(r'pdf$(r_t)$')
plt.show()
```

A qualitative good agreement is observed. We can further normalize the variable and compare it with a normal distribution of 0 mean and unit variance. That is, $X_t = (r_t - \bar{r_t})/\sigma_{r_t} \overset{?}{\sim} \mathcal{N}(0,1)$.

```python
log_returns_daily = np.array(log_returns_daily)
x_daily = (log_returns_daily-np.mean(log_returns_daily))/(np.std(log_returns_daily)) # This standarizes data
```

```python
plt.figure(figsize=(4,3))

# Counting frequencies.
counts, bin_edges = np.histogram(x_daily, bins=50, density=True)
# Calcular centros de los bins
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
plt.plot(bin_centers, counts, 'o', label=r'GOOG', linestyle = 'none', alpha = 0.5)

# Plotting N(0,1)
z_gauss = np.linspace(-4,4, 500)
normal_gauss = (1/(np.sqrt(2*np.pi)))*np.exp(-0.5*(z_gauss**2))
plt.plot(z_gauss, normal_gauss, color = 'black', linestyle = '--', label = r'$N(0,1)$')

plt.xlabel(r'$X_t = (r_t-\bar{r_t})/\sigma_{r_t}$')
plt.ylabel(r'pdf$(r_t)$')
plt.yscale('log')
plt.legend(loc = 'best', frameon = False)
plt.show()
```

We will now verify that both the mean and variance remain constant across different time windows.

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 3), sharex=True)
rolling_window_list = [10,30, 50]
log_returns_2 = np.log(data_df['GOOG'] / data_df['GOOG'].shift(1)).dropna()

colors = ["tab:blue", "tab:orange", "tab:green"]

for i in range(len(rolling_window_list)):
    # Average over time window
    rolling_mean = log_returns_2.rolling(window=rolling_window_list[i]).mean()
    rolling_var = log_returns_2.rolling(window=rolling_window_list[i]).var()

    ax[0].plot(rolling_mean, label=rf'TW {rolling_window_list[i]} days', color = colors[i], alpha = 1)
    ax[1].plot(rolling_var, label=rf'TW {rolling_window_list[i]} days', color =  colors[i], alpha = 1)
    

# Finishing up plot for mean
ax[0].axhline(log_returns_2.mean(), linestyle='--', color='black', label='Overall Mean')
ax[0].legend()
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Mean')

# Finishing up plot for variance
ax[1].axhline(log_returns_2.var(), linestyle='--', color='black', label='Overall Variance')
ax[1].legend()
ax[1].set_ylabel('Variance')
ax[1].set_xlabel('Date')

plt.tight_layout()
plt.show()
```

To easily compute the autocorrelation, we use plot_acf for visual analysis and acorr_ljungbox to statistically test for the presence of autocorrelations.

```python
# Import libraries
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
fig, ax = plt.subplots(figsize=(4, 3))  # AquÃ­ defines el tamaÃ±o
# Plot autocorrelation function
plot_acf(log_returns_2, lags=30, ax=ax)
#plot_acf(log_returns_2, lags=30, figsize=(4, 3))
plt.title("")
plt.ylim([-1.1,1.1])
plt.xlabel(r'$k$ [days] (time lag)', fontsize = 12)
plt.ylabel(r'$ \frac{1}{\sigma ^2} \mathbb{E}[(r_{t} - \bar{r})(r_{t-k}-\bar{r})]$', fontsize = 12)
plt.show()
```

```python
# Ljungâ€“Box test (up to lag 20)
ljung_box = acorr_ljungbox(log_returns_2, lags=[20], return_df=True)
print(ljung_box)
```
